```python
import pandas as pd

# Diccionario com os DataFrames que j√° temos na mem√≥ria
bases = {
    'Maio 2020': df_maio,
    'Julho 2020': df_julho,
    'Setembro 2020': df_setembro
}

# Fun√ß√£o: devolve contagem e porcentagem de nulos
def resumo_nulos(df):
    falta = df.isna().sum()
    perc  = (falta / len(df) * 100).round(1)
    return pd.DataFrame({'Ausentes': falta, '%': perc})

# Construir um √∫nico DataFrame de resumo
resumos = {}
for nome, df in bases.items():
    resumos[nome] = resumo_nulos(df)

# Exibir resumo para cada m√™s
for nome in resumos:
    print(f"\n--- {nome} ---")
    display(resumos[nome].sort_values('Ausentes', ascending=False))
```

```python
def tratar_nulos(df):
    df = df.copy()                    # evita modificar vista
    # 1) procurou_atendimento & fez_teste ‚Äì nulos‚Üí2
    df[['procurou_atendimento','fez_teste']] = (
        df[['procurou_atendimento','fez_teste']].fillna(2)
    )

    # 2) resultado_teste
    mapa_res = {48:'Positivo', 12:'Negativo', 36:'Indeterminado'}
    df['resultado_teste'] = (
        df['resultado_teste'].map(mapa_res).fillna('Sem Teste')
    )

    # 3) trabalho_remoto ‚Äì nulos‚Üí2
    df['trabalho_remoto'] = df['trabalho_remoto'].fillna(2)

    # 4) mapear 1/2 ‚Üí Sim/N√£o
    bin_cols = ['sintoma_febre','sintoma_tosse','sintoma_garganta','sintoma_falta_ar',
                'sintoma_cabeca','sintoma_peito','sintoma_nausea','sintoma_diarreia',
                'sintoma_olfato_paladar','procurou_atendimento','fez_teste','trabalho_remoto']
    df[bin_cols] = df[bin_cols].replace({1:'Sim', 2:'N√£o'})
    return df
```
